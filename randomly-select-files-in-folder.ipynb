```python
import os
import random
import shutil

duration = 12
posture = "correct"
result = "result"
path = ['correctposture/cmajor/','correctposture/fmajor/','correctposture/bmajor/']
dest_trainingdata_folder = ['result/training/correctposture/','result/training/correctposture/','result/training/correctposture/']
dest_validationdata_folder = ['result/validation/correctposture/','result/validation/correctposture/','result/validation/correctposture/']
dest_testdata_folder = ['result/test/correctposture/','result/test/correctposture/','result/test/correctposture/']

print(path[0])
#get folder array length
foldernum_i = len(path)
print(foldernum_i)
#for while statement
iteration_i = 0
j = 0

while(j < 3):
    print("j=",j)
    while(iteration_i < foldernum_i):
        print("iteration_i=",iteration_i)
        '''count file number in path'''
        #call all files in the path
        files = os.listdir(path[iteration_i])
        #call 80percents of path
        training_datanumber = int(len(files) * 0.8) #for training data
        vali_filenumber = int(len(files) * 0.5) #for training data
        test_filenumber = int(len(files))
        if(j == 0):
            a = training_datanumber #for training data
        if(j == 1):
            a = vali_filenumber
        if(j == 2):
            a = test_filenumber
        print("2a=",a)
        print("2j=",j,"a=",a)

        '''move file to another folder 1/3 -> for training data'''
        #choose n file randomly in path
        while (a > 0):
            print("a=",a,"j=",j)
            #call all files in path
            files = os.listdir(path[iteration_i]) 
            #choose 1 file randomly in path
            index = random.randrange(0, len(files))
            print(files[index])
            #store filename
            filename = files[index]
            #move file to new path
            if(j == 0):
                shutil.move(path[iteration_i]+filename, dest_trainingdata_folder[iteration_i]+filename)
            if(j == 1):
                shutil.move(path[iteration_i]+filename, dest_validationdata_folder[iteration_i]+filename)
            if(j == 2):
                shutil.move(path[iteration_i]+filename, dest_testdata_folder[iteration_i]+filename)
            a -= 1

        iteration_i += 1
    iteration_i = 0
    j += 1

```
